```python
import pandas as pd
```

```python
# Running this code will query a table in BigQuery and download
# the results to a Pandas DataFrame named `results`.
# Learn more here: https://cloud.google.com/bigquery/docs/visualize-jupyter

%%bigquery results --project shreyansh-cs777-assignments
SELECT * FROM `shreyansh-cs777-assignments.1523.businesses`
```

```python
%%bigquery business_reviews_concatenated --project shreyansh-cs777-assignments
SELECT
    DISTINCT b.business_id,
    FIRST_VALUE(b.name) OVER (PARTITION BY b.business_id ORDER BY r.date DESC) as name,
    FIRST_VALUE(b.address) OVER (PARTITION BY b.business_id ORDER BY r.date DESC) as address,
    FIRST_VALUE(b.city) OVER (PARTITION BY b.business_id ORDER BY r.date DESC) as city,
    FIRST_VALUE(b.state) OVER (PARTITION BY b.business_id ORDER BY r.date DESC) as state,
    FIRST_VALUE(b.postal_code) OVER (PARTITION BY b.business_id ORDER BY r.date DESC) as postal_code,
    FIRST_VALUE(b.stars) OVER (PARTITION BY b.business_id ORDER BY r.date DESC) as business_stars,
    FIRST_VALUE(b.review_count) OVER (PARTITION BY b.business_id ORDER BY r.date DESC) as review_count,
    STRING_AGG(r.text, ' ') OVER (PARTITION BY b.business_id) as concatenated_reviews
FROM
    `shreyansh-cs777-assignments.1523.businesses` b
JOIN
    `shreyansh-cs777-assignments.1523.reviews` r
ON
    b.business_id = r.business_id
WHERE
    b.is_open = 1 AND b.review_count > 5
```

```python
%%bigquery results_filtered --project shreyansh-cs777-assignments
SELECT *
FROM
    `shreyansh-cs777-assignments.1523.businesses` b
WHERE
    b.is_open = 1 AND b.review_count > 5
```
